```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("compact.csv")
df.head()
```

```python
df.columns
df.info()
df.describe()
```

```python
cols = ['country', 'date', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'people_vaccinated', 'population']
df_subset = df[cols]
df_subset.head()
```

```python
brazil = df_subset[df_subset['country'] == 'Brazil']
brazil_2021 = brazil[(brazil['date'] >= '2021-01-01') & (brazil['date'] <= '2021-12-31')]
```

```python
plt.plot(brazil_2021['date'], brazil_2021['new_cases'])
plt.xticks(rotation=45)
plt.title("Novos casos de COVID-19 no Brasil em 2021")
plt.show()
```
